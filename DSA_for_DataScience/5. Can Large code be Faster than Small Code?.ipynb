# Illustration with Time

**1. Constant Time Operation (O(1))**

```python
import time

def constant_operation(arr):
    start_time = time.time()
    # Accessing the first element in the array
    first_element = arr[0]
    end_time = time.time()
    print(f"Time for constant operation: {end_time - start_time} seconds")

# Example usage
arr = list(range(100000))  # Array of size 100,000
constant_operation(arr)
```

- Here acccesing the first element and printing it.
- After exceuting the code - We see as the size varies - there is very minute variation in time.
 

**2. Linear Time Operation (O(n))**

```python
import time

def linear_operation(arr):
    start_time = time.time()
    # Summing all elements in the array (linear operation)
    total = 0
    for element in arr:
        total += element
    end_time = time.time()
    print(f"Time for linear operation: {end_time - start_time} seconds")

# Example usage
arr = list(range(100000))  # Array of size 100,000
linear_operation(arr)
```

## Visualizing the Growth of Operations

```python
import matplotlib.pyplot as plt

# Input sizes
input_sizes = [1, 2, 3, 5, 10, 100]

# Constant operations (always 2)
constant_operations = [2] * len(input_sizes)
print("constant_operations :",constant_operations)

# Loop operations (equal to input size n)
loop_operations = input_sizes
print("loop_operations :", loop_operations)

# Total operations (sum of constant and loop operations)
total_operations = [constant_operations[i] + loop_operations[i] for i in range(len(input_sizes))]
print("total_operations :",total_operations)

# Plotting the results
plt.figure(figsize=(10, 6))

plt.plot(input_sizes, constant_operations, label="Constant Operations (O(1))", color='blue', marker='o')
plt.plot(input_sizes, loop_operations, label="Loop Operations (O(n))", color='green', marker='o')
plt.plot(input_sizes, total_operations, label="Total Operations", color='red', marker='o')

plt.xlabel('Input Size (n)')
plt.ylabel('Number of Operations')
plt.title('Operations Growth with Input Size')
plt.legend()
plt.grid(True)
plt.show()
```

- Total operations will be deendent on Loop operations now
    - It takes the worse case

```python
import numpy as np
import matplotlib.pyplot as plt

# Define operation count functions for different time complexities

# O(1) - Constant Time
def constant_operations_count(size):
    return 1  # Always 1 operation

# O(n) - Linear Time
def linear_operations_count(size):
    return size  # The number of operations is equal to the size

# O(n^2) - Quadratic Time
def quadratic_operations_count(size):
    return size * size  # Quadratic growth in operations

# O(log n) - Logarithmic Time
def logarithmic_operations_count(size):
    if size > 0:
        return np.log2(size)  # Logarithmic growth in operations
    return 0

# O(2^n) - Exponential Time
def exponential_operations_count(size):
    return 2 ** size  # Exponential growth in operations

# Input sizes to test
input_sizes = [10, 100, 1000, 5000, 10000, 50000]

# Smaller input sizes for exponential time (to avoid excessive runtime)
exp_input_sizes = [5, 10, 15, 20, 25, 30]

# Calculate operation counts for each complexity type
constant_ops = [constant_operations_count(size) for size in input_sizes]
linear_ops = [linear_operations_count(size) for size in input_sizes]
quadratic_ops = [quadratic_operations_count(size) for size in input_sizes]
logarithmic_ops = [logarithmic_operations_count(size) for size in input_sizes]
exponential_ops = [exponential_operations_count(size) for size in exp_input_sizes]

# Plotting the results
plt.figure(figsize=(12, 8))

plt.plot(input_sizes, constant_ops, label="Constant Time (O(1))", marker='o', color='blue')
plt.plot(input_sizes, linear_ops, label="Linear Time (O(n))", marker='o', color='green')
plt.plot(input_sizes, quadratic_ops, label="Quadratic Time (O(n^2))", marker='o', color='red')
plt.plot(input_sizes, logarithmic_ops, label="Logarithmic Time (O(log n))", marker='o', color='orange')

# Plot exponential growth on a different scale
plt.plot(exp_input_sizes, exponential_ops, label="Exponential Time (O(2^n))", marker='o', color='purple')

plt.xlabel('Input Size (n)')
plt.ylabel('Number of Operations')
plt.title('Number of Operations vs Input Size for Different Time Complexities')
plt.legend()
plt.grid(True)
plt.yscale('log')  # Using logarithmic scale to better visualize exponential growth
plt.show()
```

## Can a Longer code have better time complexity than a smaller code?

#### Example 1: Short Code but High Time Complexity (O(n^2))

```python
def sum_pairs(arr):
    total = 0
    for i in range(len(arr)):  # First loop
        for j in range(i, len(arr)):  # Second nested loop
            total += arr[i] + arr[j]
    return total

# Example usage:
arr = [1, 2, 3, 4, 5]
print(sum_pairs(arr))  # Output: sum of all pairs
```

#### Example 2: Longer Code but Better Time Complexity (O(n))

```python
def sum_elements(arr):
    # Part 1: Calculate the sum of the array
    total = 0
    for element in arr:  # Linear loop
        total += element

    # Part 2: Find the maximum element
    max_value = arr[0]
    for element in arr:  # Another linear loop
        if element > max_value:
            max_value = element

    return total, max_value

# Example usage:
arr = [1, 2, 3, 4, 5]
print(sum_elements(arr))  # Output: sum of all elements and the maximum element
```

```python
import time

# Short code with O(n^2) time complexity
def sum_pairs(arr):
    total = 0
    for i in range(len(arr)):  # First loop
        for j in range(i, len(arr)):  # Second nested loop
            total += arr[i] + arr[j]
    return total

# Longer code with O(n) time complexity
def sum_elements(arr):
    # Part 1: Calculate the sum of the array
    total = 0
    for element in arr:  # Linear loop
        total += element

    # Part 2: Find the maximum element
    max_value = arr[0]
    for element in arr:  # Another linear loop
        if element > max_value:
            max_value = element

    return total, max_value

# Creating a large input array to compare execution times
arr = list(range(1, 10000))  # Array of size 10,000

# Timing the short O(n^2) code
start_time = time.time()
sum_pairs(arr)
end_time = time.time()
print(f"Execution time for O(n^2) code: {end_time - start_time} seconds")

# Timing the longer O(n) code
start_time = time.time()
sum_elements(arr)
end_time = time.time()
print(f"Execution time for O(n) code: {end_time - start_time} seconds")
```

- Even the length of the code for O(n^2) is smaller
    - but it takes more time for exceution
    
    
- Code for O(n) - because it is a linear time space 
    - exceution time is less
    
    
    
    **It totally depends upon the type of operation the code is using**
