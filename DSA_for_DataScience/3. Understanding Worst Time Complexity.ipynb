Let’s assume we have a list of five elements: 10, 20, 30, 40, and 50. 

- Our goal is to find a specific target number in this list. 
- If that number exists, we need to return its index. 
- The index for the list is as follows: 
    - index 0 for 10, index 1 for 20, index 2 for 30, index 3 for 40, and index 4 for 50. 
    
If we find the target number, we will return its corresponding index.

For example, if we’re searching for the number 50. 

Let’s define a list as `li = [10, 20, 30, 40, 50]`, and we’ll set a target variable, which is currently 50. 

- Will create a function to check if this target is present in the list and, if so, return its index. 

We’ll call this function `target_search`, which takes a list and a target as input parameters.

```python
list1 = [10,20,30,40,50]
target = 50

def target_search(list1, target):
    for i in range(len(list1)):
        if list1[i] == target:
            return i
    return -1
```

```python
target_search(list1, target)
```


To run this function, 
- we’ll iterate through the length of the list. 
- first it's going to the first element and comparing with the target.
- If the element at the current index equals the target, 
    - we will return the index. 
- otherwise it will keep going to the next element till it finds the target value.
    
- If the target is not found, we’ll return -1. 
- Now let’s run the `target_search` function with 
    - the list and target we created above. 
    - For the target 50, it will return 4.

    - let’s say we search for 60, which is not in the list. It will return -1. 
    
    In this code, as we search for the target number, 
    - we iterate through each element one by one,

## CONCEPT OF THREE NOTATIONS :

This is the concept of the three notations discussed in data structures and algorithms regards to the time complexity. 

- The **best case** is known as the **lower bound** and is represented by **Big Omega.**

- The third case, **the average scenario**, is called **Big Theta**, while the 

- **worst case** is denoted by **Big O**.
