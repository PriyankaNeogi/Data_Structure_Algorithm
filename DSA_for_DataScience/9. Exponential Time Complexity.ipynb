# 1. Introduction to Time Complexity

Time complexity is a measure of how the time required by an algorithm increases as the size of the input (n) increases. Different types of time complexity include:

1. Linear (O(n)): Time grows linearly with input size.
2. Quadratic (O(n^2)): Time grows quadratically.
3. Exponential (O(2^n)): Time doubles with each increment in input size—leading to rapid growth.
This booklet will focus on exponential time complexity and demonstrate why it is one of the most inefficient forms of complexity.

# 2. Understanding Exponential Time Complexity

**Definition**: 
- An algorithm has exponential time complexity 
    - if the time required to execute grows at a rate proportional to a 
        - fixed power raised to the size of the input. 
        - It is generally represented as O(2^n), O(3^n), or O(k^n), 
            - where k is a constant greater than 1.

**Explanation:** 
- Imagine you have a series of switches, and 
    - each switch can be either ON or OFF. 
    - With just a few switches, you have to test all possible combinations to find the correct one. 
    - The number of combinations grows exponentially as the number of switches increases.

**Growth Comparison:**

1. Linear (O(n)): 
    - 1 switch → 1 test, 
    - 2 switches → 2 tests.
2. Quadratic (O(n^2)):
    - 1 switch → 1 test, 
    - 2 switches → 4 tests.
3. Exponential (O(2^n)): 
    - 1 switch → 2 tests, 
    - 2 switches → 4 tests,
    - 3 switches → 8 tests... 
    It grows extremely quickly.


## Exponential Time Complexity :

The **worst time complexities in the programming domain**

- Exponential time complexity. 
- Focusing on how an increase in data size impacts the magnitude of time growth.

Why is exponential time complexity considered the worst?

- We'll illustrate this with a basic tabular comparison,
    - showing how, for a data size of just 30, 
        - the number of operations can reach billions. 
    
This explains why exponential time complexity is 
- not recommended for large data sets; 
- it is only feasible for small data sets or 
- limited scenarios with very few observations.


Consider a simple example. 
- Imagine you have a lock. 
- The lock could have one digit, two digits, or three digits, and so on. 
- If the lock has only one digit, 
    - it can be either 0 or 1, resulting in two possible ways to unlock it. 
    
- For a two-digit lock, 
    - the combinations are 00, 01, 10, and 11
    -  four ways in total. 
    
- If the lock has three digits, the possible combinations are 
    - 000, 001, 010, 011, 100, 101, 110, and 111, making eight ways.

From this, we observe a pattern:  
- For 1 digit, there are \(2^1 = 2\) ways.  
- For 2 digits, there are \(2^2 = 4\) ways.  
- For 3 digits, there are \(2^3 = 8\) ways.  


In general, for a lock with \(n\) digits, 
- where each digit can take two values (0 or 1),
    - there are \(2^n\) possible combinations. 
    
For example, if there are 5 digits, 
- there are \(2^5 = 32\) ways; 
- for 10 digits, \(2^{10} = 1,024\); 
- for 20 digits, \(2^{20}\); and for 
- 30 digits, \(2^{30}\), which equals over a billion operations!

This rapid growth illustrates why exponential time complexity is undesirable for large data sets—it increases extremely quickly. 

- Moreover, it doesn't always have a base of 2. 
- For instance, if the lock can take three values 
    - (e.g., 0, 1, 2), then for a three-digit lock, 
    - there are \(3^3 = 27\) ways to unlock it.
    
Exponential time complexity generally involves operations proportional to
- \(a^n\), where \(a\) is the number of possibilities per digit.



# 3. The Classic Example: Recursive Fibonacci

A common example of exponential time complexity is the recursive calculation of Fibonacci numbers.

**Problem:** Calculate the Fibonacci sequence, where:

fib(n) = fib(n-1) + fib(n-2)

fib(0) = 0, fib(1) = 1


```python
def fibonacci(n):
    # Base cases
    if n <= 1:
        return n
    # Recursive calculation
    return fibonacci(n - 1) + fibonacci(n - 2)

# Testing the function
n = 10
for i in range(n):
    print(f"Fibonacci({i}) = {fibonacci(i)}")
```

**Time Complexity: O(2^n)**

**Explanation:** 
- The function calls itself twice for each non-base case. 
- This results in a binary tree of calls, 
    - where the number of calls doubles with each level. 
- As n increases, 
    - the number of operations grows exponentially.

Imagine you have a list of friends, and you need to decide whom to invite to your party. 

- This list could vary in size, 
    - from a single name to 10, 20, 30, or even 50 friends. 
    - For each friend, you have two choices: 
        - invite or 
        - not invite. 

Let’s explore this scenario with a simple example.

If you have just one friend, there are two possibilities:
- you either invite them (yes) or 
- don’t invite them (no). 

Now, if you have two friends, say A and B, the combinations of invitations increase:  
1. Invite both A and B  
2. Invite only A  
3. Invite only B  
4. Invite neither  

Thus, for two friends, there are four possible combinations. 

Extending this to three friends (A, B, and C), the possibilities further increase:  
1. Invite all three (A, B, C)  
2. Invite A and B  
3. Invite A and C  
4. Invite B and C  
5. Invite only A  
6. Invite only B  
7. Invite only C  
8. Invite none  

This results in eight possibilities. Observing the pattern:  
- For 1 friend: \(2^1 = 2\) possibilities  
- For 2 friends: \(2^2 = 4\) possibilities  
- For 3 friends: \(2^3 = 8\) possibilities  

If you generalize, 
- for \(n\) friends, the number of combinations is \(2^n\). 
- For instance, with 100 friends, 
    - the possibilities would be \(2^{100}\),
    - a number so large it extends into trillions or more.

Comparing this to linear time complexity,
- if you go through a list of \(n\) friends and decide for each one whether to invite them, it would take \(O(n)\) operations. 
    - For example, with 100 friends, you’d execute the decision-making process 100 times. 

In contrast, with exponential time complexity, 
- you’d have \(2^{100}\) possibilities, which is computationally infeasible.

This stark difference shows why exponential time complexity is unsuitable for large data sets. 

It works for very small datasets but 
- becomes unscalable beyond a threshold. 

For instance, at \(2^{30}\), 
- the number of operations exceeds 1 billion,
- making it impractical for real-world applications.

```python
# Greedy Algorithm:
# Function to generate all possible combinations of invitations
def invite_combinations(friends):
    # If no friends left to consider, return a single empty combination
    if not friends:
        return [[]]
    
    # Get the first friend and the rest of the list
    first_friend = friends[0]
    rest_of_friends = friends[1:]
    
    # Recursively get combinations for the remaining friends
    smaller_combinations = invite_combinations(rest_of_friends)
    
    # For each combination, create two versions: with and without the first friend
    combinations_with_first = [[first_friend] + comb for comb in smaller_combinations]
    combinations_without_first = smaller_combinations
    
    # Return all combinations
    return combinations_with_first + combinations_without_first

# List of friends - vary the number of friends
friends = ['A', 'B', 'C','D','E']

# Get all combinations
all_combinations = invite_combinations(friends)

# Display all combinations
for idx, combination in enumerate(all_combinations, start=1):
    invited = " and ".join(combination) if combination else "No one"
    print(f"Combination {idx}: Invite {invited}")
```

# 5. Exponential Growth Visualization

Let’s visualize how exponential time complexity grows compared to other types.



```python
import matplotlib.pyplot as plt
import numpy as np

# Generate input sizes
input_sizes = np.arange(1, 21)

# Calculate operations for different complexities
linear_ops = input_sizes  # O(n)
quadratic_ops = input_sizes ** 2  # O(n^2)
exponential_ops = 2 ** input_sizes  # O(2^n)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, linear_ops, label='O(n) - Linear', marker='o')
plt.plot(input_sizes, quadratic_ops, label='O(n^2) - Quadratic', marker='s')
plt.plot(input_sizes, exponential_ops, label='O(2^n) - Exponential', marker='^')

# Adding plot details
plt.title('Growth of Different Time Complexities')
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Operations')
plt.yscale('log')  # Log scale for better visualization
plt.legend()
plt.grid(True)
plt.show()
```
