### Determining Time Complexity of Code: Handling Different Sections

	1.	Identify the Different Time Complexities in Sections:
	•	Analyze each section of your code separately to determine its time complexity.
	•	For example, if one part has a loop (O(n)) and another part calls a sorting function (O(n log n)), note them down.

	2.	Focus on the Dominant Time Complexity:
	•	When multiple sections have different time complexities, the overall time complexity is dominated by the one with the highest growth rate.
	•	Example: If a code has an O(n) loop followed by an O(n^2) nested loop, the overall time complexity is O(n^2) because it grows faster as n increases.
	
3.	Combine Sections Using the Rule of Addition:

	•	If a code executes multiple parts sequentially, you sum their complexities.
	•	Example: A code with an O(n) loop followed by an O(log n) search has a combined time complexity of O(n + log n).
	•	However, since O(n) dominates O(log n), the overall complexity remains O(n).
	
4.	Handle Nested Operations Carefully:

	•	For nested loops or recursive calls, multiply their complexities.
	•	Example: An O(n) loop containing an O(log n) operation results in an overall time complexity of O(n * log n).

	5.	Choose the Dominant Complexity for Overall Analysis:

	•	When comparing different time complexities, the one that grows fastest with input size is the dominant one.
	•	Example: In a code with sections O(n), O(n^2), and O(log n), the overall complexity is O(n^2) because it becomes the largest as n grows.

	6.	Ignore Lower Order Terms for Large Inputs:
	•	When the input size n is large, lower-order terms become insignificant.
	•	Example: If a code’s complexity is O(n^2 + n), it simplifies to O(n^2), as the quadratic term dominates for large n.

- Key Takeaway: 

Always look for the **part of the code that takes the longest as input size increases,** and **that becomes the overall time complexity.** Lower time complexity sections do not affect the final complexity significantly.

## Polynomial Time Complexity

Time complexity -  **measures how the number of iterations or operations changes as the size of your dataset increases.**

For example, 
- in a linear time complexity, 
    - when you add one more element, 
        - the number of operations increases by one. 
        
- So, if you add 500 elements, 
    - 500 operations are added to the execution time. 
    
There are scenarios where 
- a slight increase in data size, 
    - like adding just two elements, 
        - can significantly increase runtime. 
        
- This is especially true in more complex codes, 
    - where polynomial or 
    - even exponential increases can occur, 
    
    which are not suitable for scalable solutions.


## Polynomial time complexity.

# 2. Linear Time Complexity - O(n)

**Concept:** The time taken grows linearly with the input size.

**Example**: If you have to call every person on a list once, the time taken grows directly with the number of people.

**Code Example**: Sum of elements in a list.


```python
def sum_elements(arr):
    total = 0
    for num in arr:
        total += num
    return total

# Example usage
numbers = [1, 2, 3, 4, 5]
print("Sum of elements:", sum_elements(numbers))
```

# 3. Quadratic Time Complexity - O(n^2)

**Concept:** The time taken grows quadratically with the input size.

**Example**: If you need to compare every person with every other person in a group (like pairing guests at a wedding), the number of comparisons grows as the square of the number of people.

**Code Example**: Finding all pairs in a list.


```python
def find_pairs(arr):
    n = len(arr)
    pairs = []
    for i in range(n):
        for j in range(i + 1, n):
            pairs.append((arr[i], arr[j]))
    return pairs

# Example usage
numbers = [1, 2, 3, 4]
print("All pairs:", find_pairs(numbers))
```

# 4. Cubic Time Complexity - O(n^3)

**Concept:** The time taken grows cubically with the input size.

**Example**: If you need to find groups of three people from a list where all three know each other, the number of possible groups grows as the cube of the number of people.

**Code Example**: Finding all triplets in a list.

```python
def find_triplets(arr):
    n = len(arr)
    triplets = []
    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                triplets.append((arr[i], arr[j], arr[k]))
    return triplets

# Example usage
numbers = [1, 2, 3, 4]
print("All triplets:", find_triplets(numbers))
```

# 5. Visualizing Polynomial Growth: Number of Operations vs. Input Size

Let's analyze how the number of operations increases with the size of the dataset. We will use linear, quadratic, and cubic examples, and plot how the number of operations grows with increasing input size.

**Visualize Growth:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate input sizes
input_sizes = np.arange(1, 101)  # Range of input sizes

# Calculate operations for each complexity (y = mx form)
m = 1  # Slope for the linear line
linear_ops = m * input_sizes  # O(n)
quadratic_ops = m * (input_sizes ** 2)  # O(n^2)
cubic_ops = m * (input_sizes ** 3)  # O(n^3)
quartic_ops = m * (input_sizes ** 4)  # O(n^4)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, linear_ops, label='O(n) - Linear (y=mx)', marker='o', linestyle='-')
plt.plot(input_sizes, quadratic_ops, label='O(n^2) - Quadratic', marker='s', linestyle='-')
plt.plot(input_sizes, cubic_ops, label='O(n^3) - Cubic', marker='^', linestyle='-')
plt.plot(input_sizes, quartic_ops, label='O(n^4) - Quartic', marker='x', linestyle='-')

# Adding plot details
plt.title('Polynomial Growth vs Input Size')
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Operations')
plt.yscale('log')  # Use log scale for better visualization
plt.legend()
plt.grid(True)
plt.show()
```
