## 2 Pointer Technique

###### In a linear search algorithm, 
- the time complexity is O(n) 
    - because we create a pointer and 
    - iterate over the list just once. 

This is the core aspect of linear search: 
- you move through the list a single time. 


### 2 Element search :
For instance, if we have a list of six elements in ascending order, 
- from 1 to 6, 
- and a target element of 8, 
- the goal is to return a pair of numbers, A and B, 
    - such that A + B equals 8.
    
##### Howthis concept is different from Linear search is that in linear search we have to find only one element (target element) - 

###### but now the scenario is different -  it's checking if the two crieria is matching or not to give the summation of the target.


Consider the list: 

[1, 2, 3, 4, 5, 6] in ascending order. 

We need to find two elements A and B such that 
- A + B equals 8. 

###### To do this using linear search, 
- I would define two variables, A and B, 
- starting both from the beginning of the list. 

With a list of six elements, 
- there are a total of 36 possible combinations of A and B. 

For each iteration of A, B will take various values. 
- For example, 
- if A starts at 1, 
- B would go through 1 to 6. 

However, none of these combinations will yield a sum of 8. 
So, after exhausting all possibilities with A set to 1, I would move A to 2 and repeat the process. 

In the next iteration - 
- will take A = 2,
- will take B = 2 not 1 as (A = 1, B = 2 already taken)



This iterative checking leads 
- to a time complexity of O(nÂ²)
- since both A and B traverse the list, 
- making it inefficient for larger datasets.



Here 
- A is traversing n times
- B is traversing n times

This scenario is not good for large cases.



```python
# Find two elements from the list which adds to give 8
list1 = [1,2,3,4,5,6]
target = 8

for i in range(len(list1)):         # O(n)
    for j in range(len(list1)):     # O(n)
        if list1[i] + list1[j] == target:     # Multiply O(n^2)
            print(list1[i], list1[j])
```

#### Scenario 2 : Sum > Target


Thereâs another scenario when the sum is greater than the target. 
-  a list of elements in increasing order, which is also sorted. (For a two pointer list has to be sorted)
- Our objective is to find two numbers A and B 
    - such that their sum is 10. 
    - Weâll follow the same structure we discussed and add some concepts.

First, Iâll define two pointers: 
- A will be 1 and B will be 15. 
- Now, I will calculate their sum, which is 16. 
- The target is 10. 
- Is my sum less than or greater than the target? 
    - This time, the sum is greater than the target. 
    
    
In the previous case, **when the sum was less than the target,**
- we moved our left pointer to the right. 

Now, since the minimum value of A is 1, 
- we canât reduce A further. 
- So, we need to move B instead (can reduce B only).

- We shift pointer B from here to the left, 
    - making it equal to 11. 
    - Now, the sum is 1 + 11, which equals 12, still greater than the target. 
    - Therefore, I need to move my right pointer again. 
    - After that, the new value for B will become 9. 
    - So, what is my sum now? 
    - Itâs 1 + 9, which equals 10. This matches my target, so I need to return the pair (1, 9).

Now, I need to find the next pair where the sum equals the target. 
- Since (1, 9) already sums to 10, 
- **I must move both pointers.** (can't move a single pointer) as the first pair is found.
- Now A will become 2 and B will become 6. My new sum is 8, 
    - which is less than 10. 
    - Since A is less than 10, 
    - I will move pointer A to the right again, making it 4, and my sum will now equal 10 with the pair (4, 6). 

If I look at the scenario now, because these two are the same, Iâll move both pointers to their next positions. 
- Whether I want to include duplicates, like (5, 5), is subjective, depending on whether I want distinct values.



### Key takeaway :
1. Sum < Target --> Move left pointer to the right (left --> +1)  - can't go back to already taken left pointers
2. Sum > Target ---> Move right pointer to the left (Right --> -1) - can't go back to already taken right pointers
3. Sum = Target ---> return two pointers ---> 
    
    To find the next pair (If you need to find the new pair - keeping one number fixed requires both the numbers to be fixed) ----> its not possible ---> we have to move both the pointers
    -  Move left pointer to right
    -  Move right pointer to left

```python
nums = [1,2,4,5,6,9,11,15]
target = 10
```

```python
# step 1 - create a left and a right pointer , intialise with index
# at any point A can never be greater than B, At max A can be equal to B
# A can never cross B, its not possible that (B =1 , A = 6) because it is same as(A =1, B = 6)
# step 2: take the sum

# 3 CONDITIONS:
# sum = Target
# sum < Target
# sum > Target
```

```python
def two_pointers(nums, target):
    
    # initialise two pointers
    left, right = 0, len(nums)-1
    
    # store all the pairs in the blank list
    res = []
    
    # condition to meet for 2 pointer
    while left < right:
        current_sum = nums[left] + nums[right]
        
        # check conditions
        # return the numbers to list, move both left and right pointers
        if current_sum == target:
            res.append([nums[left], nums[right]])
            left += 1
            right -= 1
        
        # if the list contains duplicate values (eg - '1' repeated 4 times) 
        # have to move pointers again and again
            while left < right and nums[left] == nums[left-1]:
                left +=1 
            while left < right and nums[right] == nums[right+1]:
                left +=1    
        
        # if sum is less than target, increase the left pointer
        elif current_sum < target:
            left += 1
            
        else:
            right -= 1
            
    return res
            
            
nums = [1,2,4,5,6,9,11,15]
target = 10 

two_pointers(nums, target)
```

### Reduce Time Complexity :

Now, let's discuss how the two-pointer technique reduces time complexity. 

- this technique reduces the time complexity to O(n). 

Consider a list with elements 1, 2, 3, 4, 5, and 6. 
- We start with pointer A at the beginning and pointer B at the end. 
- The total length of the list is n, and 
    - the condition is that pointer A must be less than pointer B. (A < B)

In this setup, 
- pointer A might move only to a certain point 
- while pointer B moves closer to it. 

In any scenario, 
- A plus B can only equal n at maximum. 
- Since A and B are iterating simultaneously
- one moving forward and the other moving backward
    - they effectively traverse the list in a combined total of n steps. 
    - This is why the time complexity for the two-pointer technique is O(n).
